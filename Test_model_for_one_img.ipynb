```python
from keras.models import load_model
```

```python
# Load karo apna model
model = load_model('Model1.keras')
```

```python
from keras.preprocessing import image
import numpy as np
```

```python
img1 = image.load_img('C:\\Users\\Salman\\Desktop\\University Work\\6th Samester\\Deep learning\\Cat_dog_classification\\catdogs_small\\catdogs_small\\test\\cats\\cat.1570.jpg', target_size=(150, 150))
```

```python
img1
```

```python
# Step 3: Image ko array mein convert karo
img_array1 = image.img_to_array(img1)
img_array1 = np.expand_dims(img_array1, axis=0)
img_array1 = img_array1 / 255.0
prediction = model.predict(img_array1)
print(prediction)
if prediction[0][0] > 0.5:
    print("Prediction: Dog 🐶")
else:
    print("Prediction: Cat 🐱")
```
